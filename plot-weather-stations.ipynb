# Plot all weather stations

Import the necessary libraries

```python
import json

import folium
```

## Load the data

Load in the JSON file containing the weather station information, and print out details of a single weather station

```python
with open('data/stations.json', 'rt') as f:
    details = json.load(f)

# pick out a test weather station
details['Monaghan'][3]
```

## Create the map

We create a `folium` `Map` object, and add `Marker`'s representing the weather stations.

```python
# create the map object
m = folium.Map(
    location=[54.126, -6.939],
    zoom_start=9,
)

# iterate through the list of stations, and
# add markers to the map
for county, stations in details.items():
    for s in stations:
        folium.Marker(
            [s['lat'], s['lon']],
            popup=f"{s['station']}, {county}",
        ).add_to(m)
```
