```python
import json

import folium
```

Load in the JSON file containing the weather station information

```python
with open('data/stations.json', 'rt') as f:
    details = json.load(f)

# pick out a test weather station
emyvale = details['Monaghan'][3]
emyvale
```

```python
c = 'Monaghan'
mean_lat, mean_lon = 0.0, 0.0
for s in details[c]:
    mean_lat += s['lat']
    mean_lon += s['lon']
n_stations = len(details[c])
mean_lat /= n_stations
mean_lon /= n_stations
```

## Plot the weather stations

```python
m = folium.Map(
#     location=[54.338, -6.956],
    location=[mean_lat, mean_lon],
    zoom_start=10,
)
for county, stations in details.items():
#     print(county, stations)
    for s in stations:
        folium.Marker(
            [s['lat'], s['lon']],
            popup=s['station'],
        ).add_to(m)
m
```
